## Sütun İsim Değiştirme

```python
sutunlar = ["ID", "DATE", "TIME", "LAT", "LONG", "COUNTRY", "CITY", "DISTRICT",
            "DIRECTION", "DISTANCE", "DEPTH", "XM", "MD", "SIZE", "MW", "MS", "MB"]

table = pandas.read_csv("earthquake.csv", sep = ",", header = None, skiprows = 1, names = sutunlar)
table.head(n=5)
```

--------------------------------------------------------------------------------------------

## Sütunları Silme

```python
table.drop(labels = ["ID", "XM", "MD", "MW", "MS", "MB"], axis = 1, inplace = True)
table.head(n=10)
```

## Değer Değiştirme

```python
table["DEPTH"] = table["DEPTH"] + 1
table.head(n=10)
```

```python
table["DISTRICT"] = table["DISTRICT"].str.replace("_", " ")
table.head(n=5)
```

```python
from datetime import date

def tarih_cevir(x):
    return date(int(x[:4]), int(x[5:7]), int(x[8:]))

table["DATE"] = table["DATE"].map(lambda x: tarih_cevir(x))
table.head(n=5)
```
