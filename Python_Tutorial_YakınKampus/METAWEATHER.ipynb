```python
from tkinter import *
import requests
from datetime import date
import random

arayuz = Tk()

arayuz.title("Hava Durumu")
arayuz.geometry("250x50")

etiket = Label(arayuz, text = "Şehir İsmi Girin:")
etiket.grid(column=0, row=0)

sehir = Entry(arayuz)
sehir.grid(column=1, row=0)


def rastgele_tarih_uydur():
    rastgele_yil = random.randint(2019,2021)
    rastgele_ay = random.randint(1,12)
    rastgele_gun = random.randint(1,30)
    rastgele_tarih = date(rastgele_yil, rastgele_ay, rastgele_gun)
    return rastgele_tarih.strftime("%Y/%m/%d")

def sehir_konumu_bul():
    global sehir
    yanit = requests.get("https://www.metaweather.com/api/location/search/?query={}".format(sehir.get()))
    sehir_id = yanit.json()[0]["woeid"]
    return sehir_id

def hava_durumu_bul():
    sehir_id = sehir_konumu_bul()
    rastgele_tarih = rastgele_tarih_uydur()
    yanit = requests.get("https://www.metaweather.com/api/location/{}/{}/".format(sehir_id, rastgele_tarih))
    print(" Tarih: {} \n Derece: {} \n Hava Durumu: {}".format(rastgele_tarih, yanit.json()[0]["the_temp"], yanit.json()[0]["weather_state_name"]))
    
    
    
dugme = Button(arayuz, text="Havayı Göster", command=hava_durumu_bul)
dugme.grid(column=1, row=1)

arayuz.mainloop()
```
